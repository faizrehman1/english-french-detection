```python
import pandas as pd
train_data = pd.read_csv("train.csv")
test_data = pd.read_csv("test.csv")
```

```python
labels = []

for sentence in test_data['text']:
    text_tokens = sentence.split()
    for i in range(0, len(text_tokens)): 
         isFrench = False 
         
         for j in range(0, len(french_stopwords)):    
                if(text_tokens[i] == french_stopwords[j]):  
                      labels.append("French")
                      isFrench = True
                      break
         
         if isFrench == True:
            break
          
    if isFrench == False:
            labels.append("English")

          
```

```python
test_data['target'] = labels
```

```python
test_data['Match'] = test_data.target.eq(test_data.label).astype(int)
```
