```python
import numpy as np
import pandas as pd
from langdetect import detect_langs
```

```python
df = pd.read_csv('train.csv')
```

```python
df
```

```python
df['text']
```

```python
def englishOrFrench(string):
    res = detect_langs(string)
    for item in res:
        if item.lang == "fr" or item.lang == "en":
            return item.lang
    return None

for index, row in df.iterrows():
    print(englishOrFrench(row['text'])) 
```
